### Applications for secure monitoring:
We can run this python in a highly secured zone and make sure rules prevent outgoing data flow except for a 50K status image file per day.

```
# setup RPA environment by installing PHP, Chromium web browser and RPA for Python
!apt update; apt install php; apt install chromium-browser; pip install rpa

# run r.setup() to download TagUI, modify to run Chromium browser and as root user
import rpa as r; r.setup(); r.dump(r.load('/root/.tagui/src/tagui').replace('"google-chrome"', '"chromium-browser"').replace('$headless_switch', '--no-sandbox'), '/root/.tagui/src/tagui')

# virtual display to run web browser normally (some websites reject headless mode)
!apt install xvfb; pip install pyvirtualdisplay
import pyvirtualdisplay; display = pyvirtualdisplay.Display(); display.start()
```

```
# run sample web automation workflow on RPA for Python GitHub page
 # r.close() not needed at the end since Colab will self-destruct
# virtual display.stop() not needed at the end for same reason

r.init()  # This does not download 200M because it did before in setup() above,  headless mode not needed as there is a virtual display

r.url('https://infchg.github.io/ProtecInt.html')
r.url('https://www.google.com')
r.type('//*[@name="q"]', 'information change infchg[enter]')
print(r.read('result-stats'))
r.snap('page', 'results.png')

# Reference: build on notes by https://github.com/tebelorg/RPA-Python and at https://colab.research.google.com/drive/13bQO6G_hzE1teX35a3NZ4T5K-ICFFdB5?usp=sharing including how to send Telegram notification, first look up @rpapybot to allow that
# r.telegram(1234567890, '[RPA on Colab]\ncloud automation completed')
```
